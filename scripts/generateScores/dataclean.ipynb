```python
import keras 
import pandas as pd
import numpy as np
import nltk
from spacy.lang.es.stop_words import STOP_WORDS
import spacy
from sklearn.feature_extraction.text import TfidfVectorizer

nltk.download('stopwords')

pf = pd.read_json('./subjectScores.json', encoding='utf-8')

pf.drop(['email'], axis=1, inplace=True)
```

```python
# Convertir a minusculas
pf['career'] = pf['career'].str.lower()
pf['keywords'] = [list (map(lambda x: x.lower(), y)) for y in pf['keywords']]
pf['subject'] = pf['subject'].str.lower()
pf['subject_career'] = [carreras.lower().split(",") for carreras in pf['subject_career']]
pf.head()
```

```python
# Eliminar stopwords
pf['career'] = [" ".join([word for word in carreras.split() if word not in STOP_WORDS]) for carreras in pf['career']]
pf['subject_career'] = [[" ".join([word for word in carreras.split() if word not in STOP_WORDS]) for carreras in carreras_list] for carreras_list in pf['subject_career']]
pf['subject'] = [" ".join([word for word in carreras.split() if word not in STOP_WORDS]) for carreras in pf['subject']]
pf.head()
```

```python
#Lemmatization
nlp = spacy.load("es_core_news_lg")
pf['career'] = [ " ".join([token.lemma_ for token in nlp(carreras)]) for carreras in pf['career']]
pf['keywords'] = [[ " ".join([token.lemma_ for token in nlp(carreras)]) for carreras in carreras_list] for carreras_list in pf['keywords']] 
pf['subject'] = [ " ".join([token.lemma_ for token in nlp(carreras)]) for carreras in pf['subject']]
pf['subjectcareer'] = [[ " ".join([token.lemma_ for token in nlp(carreras)]) for carreras in carreras_list] for carreras_list in pf['subject_career']]
pf.head()
```
