```python
from sklearn. neural_network import MLPRegressor
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

# carreras = [carrera for carrera in pf['career']]
# carreras = [[x] for x in carreras]

keywords = [keyword for keyword in pf['keywords']]
keywords = [[x] for x in keywords]

semester = [semester for semester in pf['semester']]
semester = [[x] for x in semester]

subject = [subject for subject in pf['subject']]
subject = [[x] for x in subject]

# subject_career = [subject for subject in pf['subject_career']]
# subject_career = [[x] for x in subject_career]

subject_semester = [subject for subject in pf['subject_semester']]
subject_semester = [[x] for x in subject_semester]

scores = [score for score in pf['subject_score']]
scores = [[x] for x in scores]

x = np.array(np.concatenate((keywords, semester, subject, subject_semester), axis=1))
y = np.array(scores)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=42)
print(x_train.shape, y_train.shape)
mlp = MLPRegressor(hidden_layer_sizes=(600, 1200, 2400), max_iter=500, alpha=0.0001,
                        solver='adam', verbose=10,  random_state=42,)
mlp.fit(x_train, y_train)
y_pred = mlp.predict(x_test)
print('MSE: ', mean_squared_error(y_test, y_pred))
print('MAE: ', mean_absolute_error(y_test, y_pred))
print('R2: ', r2_score(y_test, y_pred))
```
